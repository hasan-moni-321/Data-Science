```python
#accuracy for binary classification
accuracy = (TP+TN)/(TP+FP+FN+TN)
```

```python
#precsion for bianary classification
precision = TP/(TP+FP)
```

```python
#recall for binary classificatin
recall = TP/(TP+FN)
```

```python
#F bita
F-bita = ((1+(bita)square)*(precision*recall))/(precision+recall)

when bita=1, we call F1
when bita=2, we call F2
when bita=3, we call f3 and so on

when precision and recall both are equally important we take bita=1.
when FP(false positive) is more important/high than FN(false negative) we take bita=0 to 1. Most of the time in this 
situation bita=.5
When FN(false negative) is high, we increase bita >2. Generally in this situation we use bita=2 to 10. 

Note:sometimes bita value will change depends on situation.
```
