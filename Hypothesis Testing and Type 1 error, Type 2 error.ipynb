# Type 1 Error

Type 1 errors – often assimilated with false positives – happen in hypothesis testing when the null 
hypothesis is true but rejected

Type 1 errors have a probability of  “α” correlated to the level of confidence that you set. A test with 
a 95% confidence level means that there is a 5% chance of getting a type 1 error.

Type 1 errors can happen due to bad luck (the 5% chance has played against you) or because you didn’t 
respect the test duration and sample size initially set for your experiment.

# Type 2 Error

If type 1 errors are commonly referred to as “false positives”, type 2 errors are referred to as “false 
negatives”.                                                                                                           

 type 2 errors happen when the null hypothesis is false and you subsequently fail to reject it.
